## Нарисуйте трехмерный график двух параллельных плоскостей.

```python
import numpy as np
from pylab import *
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

```python
fig = plt.figure()
ax = Axes3D(fig)

X = np.arange(-3, 5, 0.5)
Y = np.arange(-3, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = 1*X + 3*Y
ax.plot_surface(X, Y, Z, color='green', alpha=.7)

X1 = np.arange(-3, 5, 0.5) + 2
Y1 = np.arange(-3, 5, 0.5) + 2
X1, Y1 = np.meshgrid(X1, Y1)
Z1 = 1*X1 + 3*Y1
ax.plot_surface(X1, Y1, Z1, color='blue', alpha=.8)
```

## Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
fig = plt.figure()
ax = Axes3D(fig)

X = np.arange(-3, 5, 0.5)
Y = np.arange(-3, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = X**2 + Y**2
ax.plot_surface(X, Y, Z, color='red', alpha=.8)
```

```python
fig = plt.figure()
ax = Axes3D(fig)

X = np.arange(-3, 5, 0.5)
Y = np.arange(-3, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = X**2 - Y**2
ax.plot_surface(X, Y, Z, color='yellow', alpha=.8)
```
