```python
from matplotlib.patches import Ellipse, Circle
import matplotlib.pyplot as plt
import numpy as np
import math

%matplotlib inline
```

```python
el = Circle((0,0), radius=10, color="g")
fig, ax = plt.subplots()
ax.axis('equal')
ax.add_artist(el) 
ax.set_xlim([-20, 20])
ax.set_ylim([-20, 20])
```

```python
el = Ellipse((0,0), 20, 20)
fig, ax = plt.subplots()
ax.add_artist(el) 
ax.set_xlim([-20, 20])
ax.set_ylim([-20, 20])
```
