Нарисуйте график функции:
y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
def make_chart(a, b, k):
    y = k * np.cos(x - a) + b
    return y
```

```python
x = np.linspace(-np.pi, 5*np.pi, num=200)
```

```python
a = 1.5
b = 2
k = 3
plt.plot(x, make_chart(a, b, k), color='red')
plt.grid(True)
```

```python
a = 0.5
b = 4
k = 1
plt.plot(x, make_chart(a, b, k), color='orange')
plt.grid(True)
```

```python
a = 2.5
b = 2
k = 2.3
plt.plot(x, make_chart(a, b, k), color='green')
plt.grid(True)
```
